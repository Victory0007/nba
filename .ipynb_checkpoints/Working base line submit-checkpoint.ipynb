## Finding teams that qualified for the playoffs
In order to predict teams that qualifies for the playoffs in the 2023, teams that qualified for the playoffs in the previous seasons should be flagged as 1 if qualified and 0 if not qualified(this forms the target variable)

po_data - data of teams that played in the playoffs(gotten from the nba_historical_data)

rs_raptor_with_target - dataset containing regular season data with target variable playoff(indicating if the team qualified for the playoffs that season)

```python
#The challenge is to tag teams that actually did qualify for the playoffs so ML models can actually learn the patterns required
#for qualification
#To solve this I simply extracted teams with playoff games for every year(Then aggregate by team and year) and teams that have
#play off games and flag them as 1 and those that didnt qualify and flag them as 0
##data_ = pd.read_csv("nba-data-historical.csv")
```

```python
#Teams that made it to the playoffs for each year
#po_data = playoff_data
po_data = data[data["type"] == "PO"]
print(po_data.shape)
po_data.head()
```

```python
# po_data = po_data.drop(["PIE%",'player_id', 'name_common',"type", "team_id"], axis = 1)
po_data = po_data.drop(['player_id', 'name_common',"type", "team_id"], axis = 1)
```

```python
po_data.shape
```

```python
# cols_missing_ = []
# for i in range(len(po_data.columns)):
#     if po_data[po_data.columns[i]].isnull().sum() > 0:
#         cols_missing_.append(po_data.columns[i])
#         print(f"{po_data.columns[i]}: {po_data[po_data.columns[i]].isnull().sum()} missing values")
po_data.isnull().sum()[po_data.isnull().sum().values > 0]
```

```python
#Number of columns with missing values
po_data.isnull().sum()[po_data.isnull().sum().values > 1].index.size
```

```python
po_data.drop('pos', axis = 1, inplace = True)
```

```python
po_data['play_off'] = 1
```

```python
#aggregating the data
po_data = po_data.groupby(['season', "franch_id"]).mean()
```

```python
po_data = po_data.reset_index(["season", "franch_id"])
```

```python
po_data.head()
```

```python
#Temporarily flagging every team in the regular season as 0"did not qualify for the playoffs"
nba_historical_raptor_rs["playoffs"] = 0
```

```python
nba_historical_raptor_rs.head()
```

```python
nba_historical_raptor_rs["season"].unique()
```

```python
#Flagging teams that made it to the play_offs as 1 and those that didnt as 0
k = 0
for i in nba_historical_raptor_rs["season"].unique():
    yearly_data = nba_historical_raptor_rs[nba_historical_raptor_rs["season"] == i]
    for j in range(yearly_data.shape[0]):
        if yearly_data.iloc[j]["franch_id"] in list(po_data[po_data["season"] == i]["franch_id"]):
            nba_historical_raptor_rs.iloc[k, -1] = 1
        k += 1
```

```python
nba_historical_raptor_rs["playoffs"].head()
```

```python
#Dropping rows with null values
nba_historical_raptor_rs.dropna(inplace=True)
```

```python
nba_historical_raptor_rs.head()
```

```python
rs_raptor_with_target = nba_historical_raptor_rs
```

## Nba elo
elo_data_rs: Contains elo_metrics data of players in the nba regular season

This dataset contains a playoff feature that contains ('t', 'q', 's', 'c', 'f', 'p') the indicates the kind of playoff game being played; where f indicates the teams that reached the finals in a particular nba season.

This is particularly useful when trying to predict which teams would make it to the nba finals

```python
#Exploring and Including the elo metrics to help improve the information contained in the dataset
elo_data = pd.read_csv("nba_elo.csv")
```

```python
print(elo_data.shape)
elo_data.head()
```

```python
elo_data.columns
```

```python
#To predict teams that do make it to the finals, I used a dataset of teams that has made it to the finals in the given dataset
#filling the nan values with 0
elo_data["playoff"] = elo_data["playoff"].fillna(0)
```

```python
#Teams with playoff values = 0 indicate that they didnt make it to the playoffs or the data just wasnt collected
#elo_metrics of the regular season
elo_data_rs = elo_data[elo_data["playoff"] == 0]
```

```python
#elo_data of the playoffs
elo_data_po = elo_data[elo_data["playoff"] != 0]
```

```python
elo_data_rs.head()
```

```python
missing_columns = ['carm-elo1_pre', 'carm-elo2_pre', 'carm-elo_prob1', 'carm-elo_prob2',
                   'carm-elo1_post', 'carm-elo2_post', 'raptor1_pre', 'raptor2_pre',
       'raptor_prob1', 'raptor_prob2', "importance", "total_rating"]
```

```python
elo_data_rs.isnull().sum()
```

```python
elo_data_rs.drop(missing_columns, axis=1, inplace = True)
```

```python
elo_data_rs.dropna(inplace = True)
```

```python
elo_data_rs.drop(["team2", "date"], axis=1, inplace=True)
```

```python
elo_data_rs = elo_data_rs.groupby(["season", "team1"]).mean().reset_index()
```

```python
elo_data_rs.head()
```

```python
#The various types of playoffs games
po_game_type = list(elo_data_po["playoff"].unique())
```

```python
po_game_type
```

```python
elo_data_po['finals'] = elo_data_po["playoff"].apply(lambda x: 1 if x == "f" or x == "c" else 0)
```

```python
elo_data_po.head()
```

```python
#po
elo_data_po.drop(missing_columns, axis=1, inplace=True)
```

```python
elo_data_po.dropna(inplace=True)
elo_data_po.drop(["team2", "date"], axis=1, inplace=True)
elo_data_po = elo_data_po.groupby(["season", "team1"]).mean().reset_index()
```

```python
#Any number not equal to 0 means the team made it to the final that year
elo_data_po["finals"] = elo_data_po["finals"].apply(lambda x: 1 if x > 0 else 0)
```

```python
elo_data_po.season.unique()
```

```python
#Why is this dataset only indicating teams from 2016
#The dataset information is not entirely complete
#I had to allow semifinal teams so as to give the model a good sense of teams that are most likely to qualify for the final
elo_data_po[(elo_data_po["finals"] == 1) & (elo_data_po["season"] != 2023)].head()
```

```python
elo_data_po = elo_data_po.rename(columns={"team1": "franch_id"})
```

```python
elo_data_rs = elo_data_rs.rename(columns={'team1': 'franch_id'})
```

```python
#Joining the elo metric data to the general nba data and the raptor data
#This contains all the information I need for the data model for the regular season
raptor_elo_gen_hist = pd.merge(nba_historical_raptor_rs, elo_data_rs, on=['season', 'franch_id'], how='inner')
```
